```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Require the following library:
# import matplotlib.pyplot as plt

# Receive np.array, show the image
def show_image(arr):
    fig, ax = plt.subplots(figsize=(10, 8))
    plt.imshow(np.asarray(arr))
```

```python
img = Image.new("RGB", (600,480), (0,0,255))
fig, ax = plt.subplots(figsize=(10, 8))
plt.imshow(np.asarray(img))
```

```python
for x in xrange(600):
    for y in xrange(480):
        if x > 590 and y > 470:
            print((x, y), "R:{:<3} G:{:<3} B:{:<3}".format(x/3, (x+y)/6, y/2))
        img.putpixel((x,y), (x/3, (x+y)/6, y/2))

fig, ax = plt.subplots(figsize=(10, 8))
plt.imshow(np.asarray(img))
```

```python
img.getpixel((599, 479))
i = np.asarray(img)
i.shape
i[479, 599, :]
```

It seemed that when converted as an `np.array` type it become array[y, x, (rgb)]

```python
img = Image.open("AR_Train_image/m-003-01.bmp")
show_image(img)
img.size
```

```python
img2 = np.asarray(img)

img3 = Image.fromarray(img2[0:20, :, :])
show_image(img3)

img3 = Image.fromarray(img2[50:90, :, :])
show_image(img3)

img3 = Image.fromarray(img2[:, 40:80, :])
show_image(img3)
```

```python
img4 = img.copy()
for x in xrange(img4.size[0]):
    for y in xrange(img4.size[1]):
        r, g, b = img4.getpixel((x, y))
        img4.putpixel((x, y),
                     (b, r, g))

show_image(img4)
```

```python
img5 = img.copy()
img5.convert("L")
img5.convert("LA")
```

```python
img5 = img.copy()
converted_img5 = img5.convert('L')
show_image(converted_img5)
# converted_img5.show()    # Works with Preview, not in Jupyter

img5_array = np.asarray(converted_img5, dtype=np.uint8)
show_image(Image.fromarray(img5_array[:, 0:50], 'L'))
Image.fromarray(img5_array[:, 0:50], 'L')
# Image.fromarray(img5_array[:, 0:50], 'L').show()    # Works with Preview, not in Jupyter
```

```python
img6 = Image.open("AR_Test_image/m-003-14.bmp")
show_image(img6)

box = (10, 60, 105, 120)
region = img6.crop(box)
region = region.transpose(Image.ROTATE_180)
img6.paste(region, box)
show_image(img6)
```
